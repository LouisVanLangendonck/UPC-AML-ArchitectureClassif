```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time 
import requests
from bs4 import BeautifulSoup
import urllib.request


driver = webdriver.Edge()

def scrape_style(url, style_name):
    big_iter = 780
    base = 'http://invarquit.cultura.gencat.cat/Cerca/'
    driver.get(url)
    first_building = driver.find_elements(By.CLASS_NAME, "enlacedestacado")[0]
    first_building.click()
    download_images(driver.current_url, style=style_name, big_it=big_iter)
    next_buttons = driver.find_elements(By.TAG_NAME, 'img')
    for button in next_buttons:
        if 'endavant' in button.get_attribute('src'):
            if "public" in button.get_attribute('title'):
                next_building = button
    next_building.click()
    contin=True
    while contin:
        big_iter += 1
        next_buttons = driver.find_elements(By.TAG_NAME, 'img')
        for button in next_buttons:
            if 'endavant' in button.get_attribute('src'):
                if "public" in button.get_attribute('title') and not "Off" in button.get_attribute('title'):
                    next_building = button
                elif "public" in button.get_attribute('title') and "Off" in button.get_attribute('title'):
                    download_images(driver.current_url, style=style_name, big_it=big_iter)
                    contin = False   
        if contin:
            next_building.click()
            download_images(driver.current_url, style=style_name, big_it=big_iter)    
        


def download_images(url, style, big_it):
    it = 0
    try:
        page = requests.get(url)
        soup = BeautifulSoup(page.content, "html.parser")
        candid_dupl = soup.find_all(class_ = "taulesDeDins") 
        if len(candid_dupl[0]) == 5:                    #To check if only one style is applied on the building
            img_source = driver.find_element(By.ID, "imatgeACanviar").get_attribute("src")
            urllib.request.urlretrieve(img_source, '{}/{}-{}-{}.jpg'.format(style,style, big_it, it))
            next_buttons = driver.find_elements(By.TAG_NAME, 'img')
            for button in next_buttons:
                if 'endavant' in button.get_attribute('src'):
                    if not "public" in button.get_attribute('title'):
                        next_img = button
            next_img.click()
            new_img_source = driver.find_element(By.ID, "imatgeACanviar").get_attribute("src")
            while new_img_source != img_source and it < 4:
                it += 1
                urllib.request.urlretrieve(new_img_source, '{}/{}-{}-{}.jpg'.format(style,style, big_it, it))
                next_buttons = driver.find_elements(By.TAG_NAME, 'img')
                for button in next_buttons:
                    if 'endavant' in button.get_attribute('src'):
                        if not "public" in button.get_attribute('title'):
                            next_img = button
                next_img.click()
                img_source = new_img_source
                new_img_source = driver.find_element(By.ID, "imatgeACanviar").get_attribute("src")   
    except:
        pass

URL= 'http://invarquit.cultura.gencat.cat/Cerca/Llista?consulta=MCU4K1JvbcOgbmljJQ==&index=780'
scrape_style(URL, 'romanesque')
```
